# Visualizing data

Visualizing year 2022 data in bar, scatter and line graphs.

```python
year_to_visualize = '2022'

#Color and size
sns.set_palette("Reds_r")
plt.figure(figsize=(15, 12))

#Top 50 countries
data_for_year = data[['Country Name', year_to_visualize]].sort_values(by=year_to_visualize, ascending=False)
data_for_year_top50 = data_for_year.head(50)

# Create a vertical bar chart using Seaborn
sns.barplot(y='Country Name', x=year_to_visualize, data=data_for_year_top50, orient='h')
plt.title(f'Top 50 Countries - Distribution in {year_to_visualize}')
plt.xlabel('Population')
plt.ylabel('Country Name')

plt.show()
```

```python
year_to_visualize = '2022'

# Adjust the figure size
plt.figure(figsize=(15, 12))

# Sort the data and select the top 50 countries
data_for_year = data[['Country Name', year_to_visualize]].sort_values(by=year_to_visualize, ascending=False)
data_for_year_top50 = data_for_year.head(50)

# Define colors for scatter plot and line plot
scatter_color = 'blue'  # Scatter plot color
line_color = 'black'    # Line plot color

# Create a scatter plot with joining lines
sns.scatterplot(x=year_to_visualize, y='Country Name', data=data_for_year_top50, color=scatter_color, label='Data Points')
sns.lineplot(x=year_to_visualize, y='Country Name', data=data_for_year_top50, color=line_color, label='Joining Line')

# Set title and labels
plt.title(f'Top 50 Countries - Distribution in {year_to_visualize}')
plt.xlabel('Population')
plt.ylabel('Country Name')

# Show the plot
plt.show()
```

Comparision of top 10 countries in the years 2000 and 2022

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Year to visualize
year_2000 = '2000'

# Sort the data for both years
data_for_year_2000 = data[['Country Name', year_2000]].sort_values(by=year_2000, ascending=False)
data_for_year_2022 = data[['Country Name', year_to_visualize]].sort_values(by=year_to_visualize, ascending=False)

# Select the top 10 countries for each year
top10_2000 = data_for_year_2000.head(10)
top10_2022 = data_for_year_2022.head(10)

# Set the color palette
sns.set_palette("muted")

# Set the style to whitegrid
sns.set_style("whitegrid")

# Create a vertical bar plot for 2000
plt.figure(figsize=(12, 10))
sns.barplot(x='Country Name', y=year_2000, data=top10_2000, color='blue', label='2000')

# Create a vertical bar plot for 2022
sns.barplot(x='Country Name', y=year_to_visualize, data=top10_2022, color='green', alpha=0.5, label='2022')

# Set title and labels
plt.title(f'Top 10 Countries - Comparison between {year_2000} and {year_to_visualize}')
plt.xlabel('Country Name')
plt.ylabel('Population')

# Show legend
plt.legend()

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

Change in Population of the World from 2000 to 2022

```python
import matplotlib.pyplot as plt

# Get the leading country for each year
leading_country_2000 = data.loc[data['2000'].idxmax()]['Country Name']
leading_country_2022 = data.loc[data['2022'].idxmax()]['Country Name']

# Get the population of the leading country for each year
population_2000 = data.loc[data['Country Name'] == leading_country_2000]['2000'].values[0]
population_2022 = data.loc[data['Country Name'] == leading_country_2022]['2022'].values[0]

# Calculate the percentage change
percentage_change = ((population_2022 - population_2000) / population_2000) * 100

# Years for the line graph
years = ['2000', '2022']

# Population for the line graph
population = [population_2000, population_2022]

# Plotting the line graph
plt.figure(figsize=(8, 6))
plt.plot(years, population, marker='o', color='b', linestyle='-', linewidth=2, markersize=8)
plt.title(f'Change in Population of the World from 2000 to 2022: ')
plt.xlabel('Year')
plt.ylabel('Population')
plt.grid(True)

# Show the plot
plt.show()
```

Population distribution of top 10 (Year 2000)

```python
import matplotlib.pyplot as plt

# Year to visualize
year_2000 = '2000'

# Sort the data by population in descending order
data_sorted_2000 = data.sort_values(by=year_2000, ascending=False)

# Select the top 10 countries
top_10_countries_2000 = data_sorted_2000.head(10)

# Labels for the pie chart
labels = top_10_countries_2000['Country Name']

# Sizes (population) for the pie chart
sizes = top_10_countries_2000[year_2000]

# Plotting the pie chart
plt.figure(figsize=(7, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title(f'Population Distribution of Top 10 Countries in {year_2000}')

# Show the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

Population distribution of bottom 10 (Year 2022)

```python
import matplotlib.pyplot as plt

# Year to visualize
year_2022 = '2022'

# Sort the data by population in ascending order
data_sorted_2022 = data.sort_values(by=year_2022, ascending=True)

# Select the bottom 10 countries
bottom_10_countries_2022 = data_sorted_2022.head(10)

# Labels for the pie chart
labels = bottom_10_countries_2022['Country Name']

# Sizes (population) for the pie chart
sizes = bottom_10_countries_2022[year_2022]

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title(f'Population Distribution of Bottom 10 Countries in {year_2022}')

# Show the plot
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
